```python
#targets- could be organizsm, single protein
target_query = target.search('coronavirus')#seach keyword
targets = pd.DataFrame.from_dict(target_query)#create dataframe of that search
targets
```

```python
# we are using single protein fo rfurther investigation!
#select and retrieve bioactivity data for SARS coronavirus  3C- like proteinase
```

```python
selected_target = targets.target_chembl_id[6]
selected_target
```

```python
activity = new_client.activity # what does standard_type mean?
res = activity.filter(target_chembl_id=selected_target).filter(standard_type="IC50")#find lowest standard value
```

```python
#to produce the same impact we need standard value to be least
df=pd.DataFrame.from_dict(res)
df.to_csv('bioactivity_data.csv', index=False)
```

```python
#handle missing data
df2= df[df.standard_value.notna()]
df2
```

The bioactivity data is in the IC50 unit. Compounds having values of less than 1000 nM will be considered to be active while those greater than 10,000 nM will be considered to be inactive. As for those values in between 1,000 and 10,000 nM will be referred to as intermediate.

```python
bioactivity_class=[]
for i in df2.standard_value:
  if float(i)>=10000:
    bioactivity_class.append("inactive")
  elif float(i)<=1000:
    bioactivity_class.append("active")
  else:
    bioactivity_class.append("intermediate")
```

```python
mol_cid=[]
for i in df2.molecule_chembl_id:
  mol_cid.append(i)
```

```python
canonical_smiles = []
for i in df2.canonical_smiles:
  canonical_smiles.append(i)
```

```python

standard_value = []
for i in df2.standard_value:
  standard_value.append(i)
```

```python
#combine the lists into 1 dataframe
data_tuples= list(zip(mol_cid,canonical_smiles,bioactivity_class,standard_value))
df3= pd.DataFrame(data_tuples,columns=['molecular_chembl_id','canonical_smiles','bioactivity_class','standard_value'])
```
